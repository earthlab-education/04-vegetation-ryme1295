```python
import geopandas as gpd 
```

```python
fire_url = "https://edcintl.cr.usgs.gov/downloads/sciweb1/shared/MTBS_Fire/data/composite_data/burned_area_extent_shapefile/mtbs_perimeter_data.zip"
fire_url
```

```python
fire_gdf = gpd.read_file(fire_url)
fire_gdf
```

```python
elkhorn = fire_gdf[fire_gdf['Incid_Name'] == 'ELKHORN']
elkhorn
```

```python
elkhorn_id = elkhorn[elkhorn['Event_ID'] == 'ID4550411531520230724']
elkhorn_id
```
